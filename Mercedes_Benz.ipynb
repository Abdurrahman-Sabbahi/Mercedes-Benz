# Performing Dimensionality Reduction/ Factor Analysis to reduce the dimensions of the Dataset
Applying PCA to train.csv dataframe

```python
from sklearn.decomposition import PCA

pca = PCA()
```

```python
# TO Validate i will divide my Training data into train validation and test validation data
from sklearn.model_selection import train_test_split
X_train_val,X_test_val,y_train_val,y_test_val=train_test_split(X_train,y_train,test_size=0.25,random_state=21)
```

```python
X_train_pca = pca.fit_transform(X_train_val)
X_test_pca = pca.fit_transform(X_test_val)
```

```python
print("The variance values of the PC components are : ")
print("------------------------------------------------")

print(pca.explained_variance_ratio_)
```

```python
pca2 = PCA(n_components=0.95) # Reducing the number of components to the factors that explain 95% of the variance!
```

```python
X_train_pca2 = pca2.fit_transform(X_train_val)
X_test_pca2 = pca2.fit_transform(X_test_val)

```

```python
pca2.explained_variance_ratio_
```

```python
print(X_train_pca2.shape)
print(X_test_pca2.shape)
```
